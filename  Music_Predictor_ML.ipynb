```python
# importing the data
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


music_df = pd.read_csv("music.csv")
# Input values of the dataset
X = music_df.drop(columns=["genre"]) 
# Output values of the dataset
y = music_df["genre"]

# call train test split function

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# create instance of DecisionTreeClassifier class
model = DecisionTreeClassifier()
# train model so it learns certain patterns in the data with .fit() , only use training variables
model.fit(X_train.values,y_train.values)
# ask model to predict from dataset, predict method takes a 2-D array, this is used for testing so use input values (x) for testing
predict = model.predict(X_test.values)
predict

# call accuracy_score

score = accuracy_score(y_test, predict)
score
```

```python
import joblib

# store our trained model in a file

joblib.dump(model, "music_recommender.joblib")
```

```python
# determine if joblib works

model = joblib.load("music_recommender.joblib")
# returns trained model
predict = model.predict([[21, 1]])
predict
```

```python
# show decision tree in a graph form

from sklearn import tree



tree.plot_tree(model, feature_names = ['age', 'gender'],
                    class_names = sorted(y.unique()),
                     filled = True)
```
